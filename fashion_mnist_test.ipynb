```python
import tensorflow as tf
from keras.utils import to_categorical
import numpy as np

from keras.datasets import fashion_mnist

import matplotlib.pyplot as plt
```

```python
# Load the data
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
print('Train size :', X_train.shape)
print('Test size :', X_test.shape)
```

```python
print('New Train size labels :', y_train.shape)
print('New Test size labels :', y_test.shape)
```

```python
data_description = {
    0 : 'T-shirt/top', 
    1 : 'Trouser',
    2 : 'Pullover',
    3 : 'Dress',
    4 : 'Coat',
    5 : 'Sandal',
    6 : 'Shirt',
    7 : 'Sneaker',
    8 : 'Bag',
    9 : 'Ankle boot' 
}
```

```python
# Showing some data both on Train and Test datasets

fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(12, 7))
fig.suptitle('Graphical representation of Train datasets with corresponding label !', fontsize=14, fontweight='bold')

for ax in axs.flatten():
    rnd_num = np.random.randint(low=0, high=len(X_train))
    ax.imshow(X_train[rnd_num])

    ax.set_title(data_description[y_train[rnd_num]])
    ax.set_axis_off()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(12, 7))
fig.suptitle('Graphical representation of Test datasets with corresponding label !', fontsize=14, fontweight='bold')

for ax in axs.flatten():
    rnd_num = np.random.randint(low=0, high=len(X_test))
    ax.imshow(X_train[rnd_num])

    ax.set_title(data_description[y_test[rnd_num]])
    ax.set_axis_off()
```

```python
# Scaling Data

X_train_mod = X_train.reshape(-1, 784) / 255
X_test_mod = X_test.reshape(-1, 784) / 255
```

```python
print('New Train size :', X_train_mod.shape)
print('New Test size :', X_test_mod.shape)
```

```python
# Using One-Hot Encoding

y_train_mod = to_categorical(y_train)
y_test_mod = to_categorical(y_test)
```

```python
print('New Train size labels :', y_train_mod.shape)
print('New Test size labels :', y_test_mod.shape)
```

```python
# Creating DL Model

model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(784, )),
    tf.keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

history = model.fit(X_train_mod,  y_train_mod, epochs=30, verbose=1)
```

```python
model.evaluate(X_test_mod, y_test_mod)
```

```python
# Creating DL Model

model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(784, )),
    tf.keras.layers.Dense(10, activation='sigmoid')
])

model1.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

history1 = model1.fit(X_train_mod,  y_train_mod, epochs=250, verbose=1, batch_size=1000)
```

```python
model1.evaluate(X_test_mod, y_test_mod)
```

```python
plt.figure(figsize=(12, 5))

plt.plot(history.history['accuracy'])
plt.plot(history1.history['accuracy'])

plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train Model', 'Train Model1'], loc='upper left')

plt.show()
```

```python
# Predict Labels of Model

fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(12, 7))
fig.suptitle('Graphical representation of Predicted datasets with corresponding label using Model!', fontsize=14, fontweight='bold')

for ax in axs.flatten():
    rnd_num = np.random.randint(low=0, high=len(X_test))
    ax.imshow(X_test[rnd_num])

    predicted_values = model.predict(X_test_mod[rnd_num].reshape(1, -1))
    predicted_label = tf.argmax(predicted_values, axis=1).numpy()[0]
    ax.set_title('Predicted label {}'.format(data_description[predicted_label]))
    ax.set_axis_off()
```

```python
# Predict Labels of Model1

fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(12, 7))
fig.suptitle('Graphical representation of Predicted datasets with corresponding label using Model1!', fontsize=14, fontweight='bold')

for ax in axs.flatten():
    rnd_num = np.random.randint(low=0, high=len(X_test))
    ax.imshow(X_test[rnd_num])

    predicted_values = model1.predict(X_test_mod[rnd_num].reshape(1, -1))
    predicted_label = tf.argmax(predicted_values, axis=1).numpy()[0]
    ax.set_title('Predicted label {}'.format(data_description[predicted_label]))
    ax.set_axis_off()
```

```python
# Creating DL Model

model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(784, )),
    tf.keras.layers.Dense(10, activation='sigmoid')
])

model2.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

history2 = model2.fit(X_train_mod,  y_train_mod, epochs=50, verbose=1, batch_size=1000, validation_data=(X_test_mod, y_test_mod))
```

```python
plt.figure(figsize=(12, 5))

plt.plot(history2.history['accuracy'], label='Training Accuracy')
plt.plot(history2.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])

plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```
